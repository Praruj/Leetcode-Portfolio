# Pow(x,n)
Implement `pow(x,n),` which calculates x raised to the power n (i.e., x^n).


1. Example 1:
    - Input: x = 2.00000, n = 10
    - Output: 1024.00000

2. Constraints:
    - -100.0 < x < 100.0
    - -2^31 <= n <= 2^31-1
    - n is an integer.
    - Either x is not zero or n > 0.
    - -10^4 <= x^n <= 10^4

Concept required to solve this problem is in repo location `(Leetcode-Portfolio\Math\Binary_expo.ipynb)`

```python
def mypow(x,n):
    ans=1
    while n>0:
        if n%2==1:
            ans*=x
        x*=x
        n//=2
    return ans

mypow(2.00000,10)
```

## How this works:

- if the current bit of n is 1 then multiply the base.
- square the base each step.
- shift n right (n=n//2)
- Continue units all bits of n are processed
