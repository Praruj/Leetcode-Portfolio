# Program for Square Root of Integer

# Integer Square Root in Python (DSA)

## Problem
Given a non-negative integer `n`, compute the **integer** square root:  
the largest integer `x` such that `x * x <= n`.

Examples:  
- `n = 0 -> 0`  
- `n = 1 -> 1`  
- `n = 8 -> 2`  
- `n = 10 -> 3`  
- `n = 16 -> 4`  
- `n = 27 -> 5`

---

## Naive (Linear) Approach

Idea: Try all integers from `0` upwards until `x * x` becomes greater than `n`.

```python
def int_sqrt_linear(n: int) -> int:
    if n < 0:
        raise ValueError("n must be non-negative")
    x = 0
    while x * x <= n:
        x += 1
    return x - 1


````python
def floorsqrt(n):
    start=0
    end=n
    res=1

    while start<=end:
        mid=start+(end-start)//2

        if mid*mid<=n:
            res=mid
            start=mid+1
        else:
            end=mid-1
    return res


floorsqrt(11)
````
