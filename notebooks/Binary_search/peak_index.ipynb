# 852. Peak Index in a Mountain Array
You are given an integer mountain array arr of length n where the values increase to a peak element and then decrease.

Return the index of the peak element.

Your task is to solve it in O(log(n)) time complexity.

---
# Solution
For element to be peak; it should meet following condition:
- A[p-1]<A[p]>A[p+1]

## Binary Search
1. Case 1 : mid element could be peak
2. case 2: If mid lies in left (increasing) then search in Right else, mid lies in Right then search in Left.


### Approach 1: Linear Search

```python
def mountainpeak(arr):
    n=len(arr)

    for i in range(1,n-1): # Since peak cant be at index 0 or n-1 in a valid mountain array
        if arr[i]>arr[i-1] and arr[i]>arr[i+1]:
            return i
    return -1
arr = [1, 3, 5, 4, 2]
mountainpeak(arr)
```

But, we have to solve in O(logn). 
### Approach 2 : Binary Search

```python
def mountainpeak(arr):
    left=0
    right=len(arr)-1

    while left<right:
        mid=left+(right-left)//2

        while left < right:
            mid = left + (right - left) // 2
            if arr[mid] < arr[mid + 1]:
                left = mid + 1      
            else:
                right = mid         
            
        return left  


arr = [1, 3, 5, 4, 2]
mountainpeak(arr)
```

Input: [1, 3, 5, 4, 2]

1. left=0, right=4
   mid=2, arr[2]=5 > arr[3]=4
   → right=2    [0,2]

2. left=0, right=2  
   mid=1, arr[1]=3 < arr[2]=5
   → left=2     [2,2]

3. left==right → RETURN 2
