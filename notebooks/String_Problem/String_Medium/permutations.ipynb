# 567. Permutation in String
Given two strings s1 and s2, return true if s2 contains a permutation of s1, or false otherwise.

In other words, return true if one of s1's permutations is the substring of s2.

Example 1:

- `Input:` s1 = "ab", s2 = "eidbaooo"
- `Output:` true
- `Explanation:` s2 contains one permutation of s1 ("ba").

```python
class Solution:
    def isfreqsame(self, freq1, freq2):
        for i in range(26):
            if freq1[i] != freq2[i]:
                return False
        return True

    def check_inclusion(self, s1, s2):
        # If s1 is longer, s2 cannot contain a permutation of s1
        if len(s1) > len(s2):
            return False

        freq1 = [0] * 26
        freq2 = [0] * 26

        # Frequency of characters in s1
        for char in s1:
            freq1[ord(char) - ord('a')] += 1

        # Frequency of the first window in s2
        for i in range(len(s1)):
            freq2[ord(s2[i]) - ord('a')] += 1

        # Check first window
        if self.isfreqsame(freq1, freq2):
            return True

        # Slide the window over s2
        for i in range(len(s1), len(s2)):
            # Remove the char that is leaving the window
            freq2[ord(s2[i - len(s1)]) - ord('a')] -= 1
            # Add the new char entering the window
            freq2[ord(s2[i]) - ord('a')] += 1

            # Check this window
            if self.isfreqsame(freq1, freq2):
                return True

        return False
```

```python
sol = Solution()
print(sol.check_inclusion("ab", "eidbaooo"))  
print(sol.check_inclusion("ab", "eidboaoo"))  
```
