# **Trapping Rain Water**

On using `naive method`:

- `water[i]=min(lmax, rmax) - height[i]`

```text
====== i = 2 ======
Start: lmax = 2 , rmax = 2

 LEFT LOOP:
   j = 0 | height[j] = 3 | before lmax = 2
   updated lmax = 3

   j = 1 | height[j] = 0 | before lmax = 3
   updated lmax = 3

   j = 2 | height[j] = 2 | before lmax = 3
   updated lmax = 3

 RIGHT LOOP:
   j = 2 | height[j] = 2 | before rmax = 2
   updated rmax = 2

   j = 3 | height[j] = 0 | before rmax = 2
   updated rmax = 2

   j = 4 | height[j] = 4 | before rmax = 2
   updated rmax = 4

 FINAL for i = 2 → lmax = 3 , rmax = 4
```

````python
def trap_water(height):
    n=len(height)
    total_water=0

    for i in range(n):
        lmax=height[i]
        rmax=height[i]

        #left max
        for j in range(i+1):
            lmax=max(lmax,height[j])

        
        # Right max

        for j in range(i,n):
            rmax=max(rmax,height[j])
        
        total_water+=min(lmax,rmax)-height[i]
    return total_water

height = [3, 0, 2, 0, 4]
print("Trapped Water:", trap_water(height))
````

**Complexity analysis**
- Time complexity : O(n^2) : Because we are using nested loops.
- space complexity : O(1)

**Optimal Approach**

1. Fill lmax left to right
2. Fill rmax right to left
3. water loop just reads these precomputed values. 

````python
def trap_water(height):
    n=len(height)
    # Creating size of array n. 
    lmax=[0]*n
    rmax=[0]*n

    lmax[0]=height[0]
    rmax[n-1]=height[n-1]   #“Put the last value of height into the last position of rmax”
    
    for i in range(1,n):
        lmax[i]=max(lmax[i-1],height[i])  #“At index i, store the maximum so far from left”

    for i in range(n-2,-1,-1):
        rmax[i]=max(rmax[i+1],height[i])

    total_water=0
    for i in range(n):
        total_water+=min(lmax[i],rmax[i])-height[i]
    return total_water

height = [3, 0, 2, 0, 4]
print("Trapped Water:", trap_water(height))
````

**Complexity Analysis**
- Time complexity: Each loop runs in n times so `O(n)`.
- Space complexity :  Each array size is n so `O(n)` space. 

So we traded space with time. 

**Optimal approach 1 : Two pointers**

````python
def trap_water(height):
    n=len(height)
    ans=0
    l=0
    r=n-1
    lmax,rmax=0,0

    while l<r:
        lmax=max(lmax,height[l])
        rmax=max(rmax,height[r])
        
        if lmax<rmax:
            ans+=lmax-height[l]
            l+=1
        else:
            ans+=rmax-height[r]
            r-=1
    return ans

height = [3, 0, 2, 0, 4]
print("Trapped Water:", trap_water(height))
````

**Complexity Analysis**

- Time complexity : O(n)
- Space Complexity : O(1)
