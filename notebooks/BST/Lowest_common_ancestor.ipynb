# Lowest Common Ancestor in Binary Tree

The LCA of two nodes `P` and `q` in a binary tree is :  
- The lowest node in the tree has both p and q as descendants. 

 *Note*
- A node can be descendant of itself.
- Therefore a node can also be ancestor of itself. 


**Recursive Intutuion**

1. Do i find p or q on my left ?
2. Do i find p pr q on my right ?

```python
class Node:

    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None

def lca(root,p,q):

    if root is None:
        return None


    # If current node is p or q
    if root==p or root==q:
        return root      #assuming this is LCA

    left_lca=lca(root.left,p,q)
    right_lca=lca(root.right,p,q)


    # If both sides returned a node,
    # current root is the LCA
    if left_lca and right_lca:
        return root
    
    # Otherwise return whichever side found something
    return left_lca if left_lca else right_lca
```

*Calling the function*

```python
root = Node(3)

root.left = Node(5)
root.right = Node(1)

root.left.left = Node(6)
root.left.right = Node(2)

root.right.left = Node(0)
root.right.right = Node(8)

root.left.right.left = Node(7)
root.left.right.right = Node(4)


answer = lca(root, root.left, root.left.right.right)

print("LCA:", answer.data)
```

### Complexity Analysis:

1. Time complexity: $O(n)$ : where `n` is number of nodes in binary tree.
2. Space complexity: $O(h)$ : where `h` is height of tree. The space comes from recursive call stack.
