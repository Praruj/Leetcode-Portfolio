# Binary Trees

A **binary tree** is a hierarchical, non-linear data structure that starts with a single **root node**. 

- Every node can have **at most two children** (often called the **left child** and **right child**).
- Each node (except the root) has exactly **one parent**.
- Nodes with no children are called **leaves**.


This "binary" restriction (maximum of 2 children per node) distinguishes binary trees from general trees, where a node can have any number of children.

---

**Build a Binary Tree**

Pre order Sequence

left  ->  Right -> Right

```python
class Node:

    def __init__(self,data):
        self.data=data
        self.left=None          # pointer to left child
        self.right=None         # pointer to right child
        
def print_preorder(root):

    if root:
        print(root.data,end=" ")

        print_preorder(root.left)

        print_preorder(root.right)

root = Node(10)
root.left = Node(5)
root.right = Node(15)

print_preorder(root) 
```
