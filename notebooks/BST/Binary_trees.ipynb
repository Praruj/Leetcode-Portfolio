# Binary Trees

A **binary tree** is a hierarchical, non-linear data structure that starts with a single **root node**. 

- Every node can have **at most two children** (often called the **left child** and **right child**).
- Each node (except the root) has exactly **one parent**.
- Nodes with no children are called **leaves**.


This "binary" restriction (maximum of 2 children per node) distinguishes binary trees from general trees, where a node can have any number of children.

- `Height` =    Levels of tree, Maximum Depth
- `subtree` = Small Part of main tree

---

**Build a Binary Tree**

Pre order Traversal
Preorder: Root → Left → Right

````python
class Node:

    def __init__(self,data):
        self.data=data
        self.left=None          # pointer to left child
        self.right=None         # pointer to right child
        
def print_preorder(root):

    if root:
        print(root.data,end=" ")

        print_preorder(root.left)

        print_preorder(root.right)

root = Node(10)
root.left = Node(5)
root.right = Node(15)

print("preorder")
print_preorder(root) 
````

---

## Inorder Traversal
Inorder: Left → Root → Right

````python
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data,end=" ")
        inorder(root.right)
    
print("Inorder:")
inorder(root) 
````

## Post order Traversal

Postorder: Left → Right → Root

````python
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data,end=" ")
print("\nPostorder:")
postorder(root) 
````

## Level Order Traversal : `Iterative`

- Time complexity : $O(n)$
- Space complexity : $O(n)$

    1. In worst case, queue stores an entire level of the tree.
    2. for a complete binary tree, the last level can have about $n/2$ nodes.
    3. so the queue can grow to $O(n)$ in size.

````python
def levelorder(root):
    if not root:
        return
    
    q=[root]

    while q:
        node=q.pop(0)
        print(node.data,end=" ")

        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)

print("level order")
levelorder(root)
````

# Height of Tree

Maximum level of binary tree or Maximum distance from root to leaf

````python
def height(root):
    if root is None:
        return 0
    left_ht=height(root.left)
    right_ht=height(root.right)

    return max(left_ht,right_ht)+1
height(root)
````

> **Note**
>
> ```python
> if root:
>     return 0
> ```
>
> **Meaning:**  
> If `root` is **truthy**, return `0`.
>
> In the context of a binary tree, this effectively means:  
> “If the node exists, stop and return 0.”
>
> However, this is usually **incorrect** for tree height logic.  
> You typically want to check:
>
> ```python
> if root is None:
>     return 0
> ```
>
> Which means:  
> “If the node does *not* exist, return 0.”

--- 
Time complexity : $O(N)$

## Count of Nodes

Total count of all right subtrees and left subtrees +1

````python
def count(root):
    if root is None:
        return 0
    left_count=count(root.left)
    right_count=count(root.right)

    return left_count+right_count+1

count(root)
````

### Sum of Nodes
**Time complexity : $O(n)$**

**Time complexity : $O(n)$**

Traversing nodes of tree in every n times.

````python
def sum(root):

    if root is None:
        return 0

    left_sum=sum(root.left)
    right_sum=sum(root.right)

    return left_sum+right_sum+root.data

sum(root)
````
