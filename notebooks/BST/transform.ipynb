# Transform to Sum Tree

- `node=node value + left subtree + right subtree`

```python
class Node:

    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None


def sumtree(root):

    if root is None:
        return 0 

    left_sum=sumtree(root.left)
    right_sum=sumtree(root.right)
    root.data+=left_sum+right_sum

    return root.data
```

```python
root = Node(10)
root.left = Node(5)
root.right = Node(3)

sumtree(root)

print(root.data)        
print(root.left.data)   
print(root.right.data)  
```

## Complexity Analysis:

- Time complexity : $O(n)$, each node is visited exactly once and at each node we do recursive calls.
- Space complexity : $O(n)$, in skewed tree, height =n
