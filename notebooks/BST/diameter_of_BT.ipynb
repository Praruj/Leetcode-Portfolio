# Diameter of Binary Tree

- Length of longest path between any two nodes of binary tree. Those two nodes are leaves. 

    1. Diameter through root node. -> `length=left height + right height`
    2. Diameter through right subtree 
    3. Diameter through left subtree


```python
class Node:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None

def height(node):
    if node is None:
        return 0
    return 1 + max(height(node.left), height(node.right))

def diameter(root):
    if root is None:
        return 0
    left_diameter=diameter(root.left)
    right_diameter=diameter(root.right)
    current_diameter=height(root.left)+height(root.right)

    return max(current_diameter, left_diameter, right_diameter)

root=Node(1)
root.left=Node(2)
root.right=Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

print(diameter(root))
```

## **Complexity Analysis**

- Time complexity : $O(N^2)$  , because we call height inisde diameter. In large tree, we will be visiting same nodes over and over again.
