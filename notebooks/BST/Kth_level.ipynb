# Kth Level of Binary Tree

-  To solve the problem we can use level order traversal

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None

def kthlevel(root,k):

    if root is None:
        return 

    if k==1:
        print(root.data)
        return

    kthlevel(root.left,k-1)
    kthlevel(root.right,k-1)

root=Node(1)
root.left=Node(2)
root.right=Node(3)
root.left.left=Node(4)
root.left.right=Node(5)

kthlevel(root,2)
```

##  Complexity Analysis

- Time complexity : $O(N)$ -> In worst case, we have to visit every node in the tree.
- Space complexity : $O(h)$ -> where h = height of tree.

## another Method

```python
from collections import deque

def kth_level(root, k):
    if not root:
        return []
    
    queue = deque([root])
    level = 1
    
    while queue:
        size = len(queue)
        
        if level == k:
            return [node.val for node in queue]
        
        for _ in range(size):
            node = queue.popleft()
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        
        level += 1
    
    return []
```
