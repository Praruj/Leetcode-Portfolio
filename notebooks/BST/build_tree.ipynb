# Build Binary Tree from Preorder and Inorder Traversal

## Problem Statement

Given two traversal arrays:

- **Preorder**: Root → Left → Right
- **Inorder**: Left → Root → Right

Construct the original binary tree.

---

## Example

```text
preorder = [3, 9, 20, 15, 7]
inorder  = [9, 3, 15, 20, 7]

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def search(inorder, left, right, value):
    for i in range(left, right + 1):
        if inorder[i] == value:
            return i
    return -1


def buildtree(preorder, inorder, left, right, preidx):
    if left > right:
        return None

    root_val = preorder[preidx[0]]
    root = Node(root_val)
    preidx[0] += 1

    if left == right:
        return root

    inidx = search(inorder, left, right, root_val)

    root.left = buildtree(preorder, inorder, left, inidx - 1, preidx)
    root.right = buildtree(preorder, inorder, inidx + 1, right, preidx)

    return root



preorder = [3, 9, 20, 15, 7]
inorder = [9, 3, 15, 20, 7]

preidx = [0]  

root = buildtree(preorder, inorder, 0, len(inorder) - 1, preidx)

print(root.data)
````

## Complexity Analysis


- Time complexity :$O(n^2)$
- space complexity : $O(n)$   -> due to recursion stack and tree storage
