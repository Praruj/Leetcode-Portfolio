# Binary Tree Paths

```Text
- Check every nodes wether it is leaf node or not. [ in leaf node, node.left==Null and node.right==Null]

````python
class Node:

    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None


def all_paths(root,path,ans):

    if root is None:
        return

    path=path+[root.data]


    if root.left is None and root.right is None:
        ans.append(path)
        return

    all_paths(root.left,path,ans)
    all_paths(root.right,path,ans)

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.right = Node(5)

ans = []
all_paths(root, [], ans)

print(ans)
````

## Complexity Analysis

- Time complexity : $O(n^2)$
- space complexity: $O(l*h)$
