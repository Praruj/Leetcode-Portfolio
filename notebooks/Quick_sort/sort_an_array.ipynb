# Sort an Array with 0s, 1s and 2s

#### **Brute force Approach**

```python
def sortcolors(nums):
    return sorted(nums)
nums = [2,0,2,1,1,0]
print(sortcolors(nums))
```

*Notes:*

- Use `sorted()` when:
    1. to Keep the list unchanged, working with tuples, dictionaries or other iterables, when we want to assin the stored result to a varable for further use.

- use `.sort()` when:
    1. When we are certain that we wont need an original order, while working with large list and performance matters. 

#### **Optimized Solution**

```python
# Store the count of 0s, 1s, 2s
# write 0 count0 times
# write 1 count1 times
# write 2 count2 times

def sortcolors(nums):
    n=len(nums)
    count0=0
    count1=0
    count2=0

    for i in range(n):
        if nums[i]==0:
            count0+=1
        elif nums[i]==1:
            count1+=1
        else:
            count2+=1
    
    idx=0
    for i in range(count0):
        nums[idx]=0
        idx+=1
    
    for i in range(count1):
        nums[idx]=1
        idx+=1

    for i in range(count0):
        nums[idx]=2
        idx+=1

    return nums


nums = [2,0,2,1,1,0]
print(sortcolors(nums))
```

#### **Optimal Solution**

Dutch National Flag algorithm : It will help us to sort an array in O(N), time complexity with single pass. 

Uses three pointers: Low, mid and High

Three region in array: 

[ 0s region | 1s region | unknown | 2s region ]

```python
def sortcolors(nums):

    n=len(nums)
    mid,high,low=0,n-1,0

    while mid<=high:

        if nums[mid]==0:
            nums[low],nums[mid]=nums[mid],nums[low]
            mid+=1
            low+=1
        
        elif  nums[mid]==1:
            mid+=1
        
        else:
            nums[high],nums[mid]=nums[mid],nums[high]
            high-=1

```

###  Complexity Analysis

- Time complexity : O(n)
- Space Complexity : O(1)
