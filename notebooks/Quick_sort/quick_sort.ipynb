# QuicK Sort Agorithm

It is based on pivot and partition.  pivot is special index, around its when sort our array. We always pick last element as pivot element. And then, we do partition of array around that pivot.  

```python
def quicksort(nums,start,end):
    if start<=end:
        pividx=partition(nums,start,end)
        quicksort(nums,start, pividx-1)
        quicksort(nums,pividx+1,end )

def partition(nums,start,end):
    idx=start-1
    pivot=nums[end]
    for j in range(start,end):
        if nums[j]<=pivot:
            idx+=1
            nums[idx], nums[j] = nums[j], nums[idx]
    nums[idx+1],nums[end]=nums[end],nums[idx+1]
    return idx+1
nums = [10, 7, 8, 9, 1, 5]
quicksort(nums, 0, len(nums) - 1)
print(nums)
```

# Complexity Analysis

- Time complexity : O(nlogn)
    Pivot divides the array into two nearly equal halves every time.
    
    In worst case, it can become O(n^2), since pivot is always smallest or largest element. 

- Space complexity:
    1. Best case : O(logn)
    2. Worst case : O(n)
    Because recursion depth depends on partition balance.

# Merge two sorted arrays

```python
def merge_sorted_arrays(arr1, arr2):
    i = j = 0
    merged = []

    while i < len(arr1) and j < len(arr2):
        if arr1[i] <= arr2[j]:
            merged.append(arr1[i])
            i += 1
        else:
            merged.append(arr2[j])
            j += 1

    # add remaining elements
    merged.extend(arr1[i:])
    merged.extend(arr2[j:])

    return merged


arr1 = [1, 3, 5, 7]
arr2 = [2, 4, 6, 8]

print(merge_sorted_arrays(arr1, arr2))
```
