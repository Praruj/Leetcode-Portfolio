# Permutations of array 

- Possible arrangements of array. If size of array is n then the permutation of array is equal to n! (n-factorial). 

```python
def permutations(nums,idx,ans):
    n=len(nums)
    if idx==n:
        ans.append(nums.copy())  # list are mutable so without copy every answer will point to the same list in memory. 
        return 
    
    for i in range(idx,n):
        nums[idx],nums[i]=nums[i],nums[idx]
        permutations(nums,idx+1,ans)
        nums[idx],nums[i]=nums[i],nums[idx]

nums = [1,2,3]
ans = []
permutations(nums, 0, ans)
print(ans)
```

### Complexity Analysis

- Time Complexity :  O(n!) -> Because there are n! permutations.
