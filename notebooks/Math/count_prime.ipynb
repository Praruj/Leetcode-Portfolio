# Seive of Erastosthenes

The Sieve of Eratosthenes is a fast way to find all prime numbers up to n. A prime number is a number greater than 1 that has only two factors: 1 and itself.
Example: `2, 3, 5, 7, 11, 13...`

```python
def countprimes(n):

    if n<2:
        return 0
    
    isprime=[True]*n
    isprime[0]=isprime[1]=False  #0 and 1 are not prime.

    count=0
    for i in range(2, int(n**0.5) + 1):
        if isprime[i]:                  #If i is still prime → remove its multiples.
            for j in range(i*i,n,i):
                isprime[j]=False       #Mark multiples of i as not prime.
    
    count=sum(isprime)
    return count

countprimes(30)
```
