# 2D array in Python

It is essentially a list of lists - each elements of main list is another list representing a row.

**Creating a 2D array**

```python
array=[
    [1,2,4],
    [3,5,6],
    [7,8,9]
]

array
```

**Accessing Elements**
- Syntax : `arr[row][column]`

```python
array[0][1]
```

**Looping through 2d array**

```python
for row in array:
    for value in row:
        print(value,end=" ")
    print()
```

**Linear search in 2D array**

```python
target=9
for i in range(len(array)):
    for j in range(len(array[i])):
        if array[i][j]==target:
            print(f"found at row {i} and column {j}.")
```

#### *Time  Complexity:* 
O(n) on worst case
Linear search works on unsorted lists, unlike binary search which requires sorting.
