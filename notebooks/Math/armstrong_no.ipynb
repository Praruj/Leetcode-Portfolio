# Armstrong Number

Armstrong number is a number that is equal to sum of cubes of its digits. 

For example:
1. 153
    - 1^3+5^3+3^3= 1+125+27=153

```python
def isarmstrong(n):
    copyn=n
    sumofcubes=0
    while n!=0:
        dig=n%10
        sumofcubes+=(dig*dig*dig)
        n=n//10
    
    return sumofcubes==copyn

n = 153
if isarmstrong(n):
    print("is an armstrong number")
else:
    print("not an armstrong number")
```

✅ Key idea

Number → (%10 machine) → last digit

Number → (//10 machine) → number without last digit

**Same Algorithm can be used in Different ways**

*Reverse a number*

reverse = reverse * 10 + digit shifts digits left and adds the new digit at the end.

```python
n=1234
reverse=0
while n!=0:
    digit=n%10
    reverse=reverse*10+digit
    n=n//10
print(reverse)
```

*Palindrome Number*

```python
n=121
copyn=n
reverse=0

while n!=0:
    digit=n%10
    reverse=reverse*10+digit
    n=n//10

if reverse==copyn:
    print("it is a palindrome")
else:
    print("it is not palindrome")
```
