# GCD
Greatest common Divisor is the largest number that duivides two or more numbers exactly (without leaving a remainder).

## Example 1

**Numbers:** 20 and 28

**Divisors of 20**
- 1
- 2
- 4
- 5
- 10
- 20

**Divisors of 28**
- 1
- 2
- 4
- 7
- 14
- 28

**Common divisors**
- 1
- 2
- 4

**The largest one is 4.**

 **GCD(20, 28) = 4**

```python
a=20
b=28
gcd=1

for i in range(1,min(a,b)+1):
    if a%i==0 and b%i==0:
        gcd=i
print(gcd)
```

“If i divides both a and b with no remainder, then i is a common divisor.”

**Euclidean Algorithm**

The more efficient algorithm is `Euclidean Algorithm`.

It is fast method to find the GCD of two numbers. 
- gcd(a,b)=gcd(b,a%b)

We keep replacing numbers until the remainder becomes 0. 
The last non zero number is GCD.

```python
a = 20
b = 28

while b != 0:
    a, b = b, a % b

print(a)
```

OR

```python
def gcd(a,b):
    while a>0 and b>0:
        if a>b:
            a=a%b
        else:
            b=b%a
    if a==0:
        return b
    else : 
        return a
    
gcd(20,28)
```

## OR

```python
def gcdrec(a,b):
    if b==0:
        return a
    return gcdrec(b,a%b)
```

## LCM

Least common multiple. If we have two values `a`, `b`:

- `a * b = gcd(a, b) * LCM(a, b)`

```python
def gcdrec(a,b):
    if b==0:
        return a
    return gcdrec(b,a%b)

def lcm(a,b):
    gcd=gcdrec(a,b)
    return (a*b)/gcd

lcm(20,28)
```
