# Modulo Arithmetic Properties

Modulo arithmetic works with **remainders after division**.

Example:

$$
17 \mod 5 = 2
$$

---

## 1. Addition Property

$$
(a + b) \mod n = ((a \mod n) + (b \mod n)) \mod n
$$

Example:

$$
(17 + 12) \mod 5 = (2 + 2) \mod 5 = 4
$$

---

## 2. Subtraction Property

$$
(a - b) \mod n = ((a \mod n) - (b \mod n)) \mod n
$$

Example:

$$
(20 - 7) \mod 6 = (2 - 1) \mod 6 = 1
$$

---

## 3. Multiplication Property

$$
(a \times b) \mod n = ((a \mod n) \times (b \mod n)) \mod n
$$

Example:

$$
(13 \times 8) \mod 5 = (3 \times 3) \mod 5 = 4
$$

---

## 4. Power Property

$$
a^k \mod n = (a \mod n)^k \mod n
$$

Example:

$$
7^2 \mod 5 = 2^2 = 4
$$

---

## 5. Congruence Notation

$$
a \equiv b \pmod n
$$

Meaning: **a and b leave the same remainder when divided by n**.

Example:

$$
17 \equiv 2 \pmod 5
$$

# **Fizz Buzz** 

```python
def fizzbuzz(n):
    result=[]

    for i in range(1,n+1):
        if i%3==0 and i%5==0:
            result.append("Fizzbuzz")
        
        elif i%3==0:
            result.append("fizz")
        
        elif i%5==0:
            result.append("buzz")
        
        else:
            result.append(str(i))
    return result

n=5
fizzbuzz(n)
```

**Given an integer array nums and an integer k, return the number of subarrays whose sum is divisible by k.**

```python
def subarray(nums,k):
    count=0
    prefix=0

    mod_count={0:1}

    for num in nums:
        prefix+=num
        mod=prefix%k

        if mod in mod_count:
            count+=mod_count[mod]

        mod_count[mod]=mod_count.get(mod,0)+1
    
    return count

nums=[4,5,0,-2,-3,1]
k = 5
subarray(nums,k)
```

```python
def subarray(nums, k):
    count = 0
    prefix = 0
    mod = {0: 1}

    for num in nums:
        prefix += num
        remainder = prefix % k

        if remainder in mod:
            count += mod[remainder]
        else:
            mod[remainder] = 0

        mod[remainder] += 1

    return count


nums = [4,5,0,-2,-3,1]
k = 5
print(subarray(nums, k))
```
