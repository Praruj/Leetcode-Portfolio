# Linked list

Linked list is a clinear data structure where elements are connected using pointers that store the address of the next element.

Instead of storing elements next to each other in memory (like arrays), each element
points to the **next element** in the list.

## Simple Visualization

[1] → [2] → [3] → [4] → NULL

- Each box is called a **node**
- Each node stores:
  - **Data**
  - **Pointer (reference)** to the next node

## Node Structure

Each node contains two parts:

1. **Data** – the value stored in the node
2. **Next Pointer** – the address of the next node

Example:

| Data | Next |
|-----|-----|
| 1 | address of node 2 |

## First Node

The first node is called the **Head**.

Head → [1] → [2] → [3] → [4] → NULL

## Last Node

The last node points to **NULL**, meaning the list ends.

## Why Use Linked Lists?

Linked lists are useful when:

- You need **dynamic size**
- You often **insert or delete elements**

## Advantages

- Dynamic size (can grow or shrink)
- Easy insertion and deletion
- No need for continuous memory

## Disadvantages

- Uses extra memory for pointers
- Slower access than arrays
- Cannot access elements directly (must traverse)

## Basic Operations

Common linked list operations:

- Insert a node
- Delete a node
- Traverse (visit all nodes)
- Search for a value

## Implementation from Scratch

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class LinkedList:
    def __init__(self):
        self.head=None

    def append(self,data):
        new_node=Node(data)

        if self.head is None:
            self.head=new_node
            return
        current=self.head
        while current.next:
            current=current.next
        current.next=new_node

    def prepend(self,data):
        new_node=Node(data)
        new_node.next=self.head
        self.head=new_node

    def delete(self,key):
        current=self.head

        if current and current.data==key:
            self.head=current.next
            return
        prev=None

        while current and current.data!=key:
            prev=current
            current=current.next

        if current is None:
            return
        
        prev.next=current.next

    def display(self):
        current=self.head
        while current:
            print(current.data,end="->")
            current=current.next
        print("None")

l1 = LinkedList()

l1.append(1)
l1.append(2)
l1.append(3)
l1.prepend(0)

l1.display()

l1.delete(2)
l1.display()
```
