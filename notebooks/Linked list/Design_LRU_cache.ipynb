# Design a LRU cache

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.prev = None
        self.next = None

class LRUCache:
    def __init__(self, capacity: int):
        self.capacity = capacity
        self.cache = {}  # key -> node
        # Dummy nodes for easy removal at head and tail
        self.head = Node(0, 0)
        self.tail = Node(0, 0)
        self.head.next = self.tail
        self.tail.prev = self.head

    # Move the node to the front (most recent)
    def _move_to_front(self, node: Node):
        # Remove node from its current position
        node.prev.next = node.next
        node.next.prev = node.prev
        # Add node right after the head (most recent)
        node.next = self.head.next
        node.prev = self.head
        self.head.next.prev = node
        self.head.next = node

    # Remove the least recently used node (from the tail)
    def _remove_from_tail(self):
        # Tail's previous node is the LRU
        lru = self.tail.prev
        self._remove_node(lru)
        return lru

    # Remove a node from the doubly linked list
    def _remove_node(self, node: Node):
        node.prev.next = node.next
        node.next.prev = node.prev

    def get(self, key: int) -> int:
        if key in self.cache:
            node = self.cache[key]
            self._move_to_front(node)
            return node.value
        return -1

    def put(self, key: int, value: int) -> None:
        if key in self.cache:
            # Update the value and move to the front
            node = self.cache[key]
            node.value = value
            self._move_to_front(node)
        else:
            # Add the new node to the cache
            if len(self.cache) >= self.capacity:
                # Cache is full, remove the LRU from the linked list
                lru = self._remove_from_tail()
                # Also remove from the cache
                del self.cache[lru.key]
            
            # Create a new node and add it to the front
            node = Node(key, value)
            self.cache[key] = node
            self._move_to_front(node)
```

## Complexity Analysis:

- Time complexity : $O(1)$
- Space complexity : $O(1)$
