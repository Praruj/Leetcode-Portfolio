# Last Stone Weight


You’re given an array of stone weights. Each turn:

- Pick the two heaviest stones
- Smash them:
- If equal → both destroyed
- If different → push back the difference
- Repeat until ≤ 1 stone remains
- Return the last stone (or 0)

---

**Brute force**

```python
def laststone(stones):
    while len(stones)>1:
        stones.sort()

        y=stones.pop()
        x=stones.pop()


        if y!=x:
            stones.append(y-x)
    return stones[0] if stones else 0

stones=[2,7,4,1,8,1]
laststone(stones)
```
