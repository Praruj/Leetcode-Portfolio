# Implement 1 - nearest Neighbour (1NN) using numpy

```TEXT
Given:

- A training dataset X_train of shape (n_train, d)
- Training labels y_train of shape (n_train,)
- A test dataset X_test of shape (n_test, d)

Implement a function that predicts the label of each test point using 1-Nearest Neighbor classification with Euclidean distance.


```

X_train = np.array([

**Vectorized NumPY solution**

````python
import numpy as np

def predict_1nn(X_train,y_test,X_test):
    # Compute pairwise squared Euclidean distances
    distances=(
        np.sum(X_test**2,axis=1,keepdims=True)
        + np.sum(X_train**2,axis=1)
        -2*X_test @X_train.T
    )
    
    nearest_idx=np.argmin(distances,axis=1)
    return y_train[nearest_idx]
````

### Nearest Neighbor solution

````python
import numpy as np

def one_nn(X_train, y_train, X_test):
    X_train = np.asarray(X_train)
    y_train = np.asarray(y_train)
    X_test = np.asarray(X_test)

    # Pairwise squared Euclidean distances
    dists = ((X_test[:, None, :] - X_train[None, :, :]) ** 2).sum(axis=2)

    # Nearest training point for each test point
    nearest = np.argmin(dists, axis=1)

    return y_train[nearest]
````
