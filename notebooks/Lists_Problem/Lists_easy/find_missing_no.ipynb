# Find Missing Number

You’re given an array of size n containing distinct numbers from 0 to n, but one number is missing.

`Missing = Expected Sum - Actual Sum`

```python
def missingnumber(nums):
    n=len(nums)
    expected=n*(n+1)//2
    actual=sum(nums)
    return expected-actual

nums = [3, 0, 1]
missingnumber(nums)
```

## Complexity Analysis:

- Time complexity : $O(n)$
- Space complexity : $O(1)$
