# Gas Station

- Gas = [1,2,4,5,9] & cost=[3,4,1,10,1]

Find the starting gas station index where you can travel around the entire circuit once.
If it's impossible, return -1.

```python
def gas_station(gas,cost):
    if sum(gas)<sum(cost):
        return -1

    start=0
    tank=0

    for i in range(len(gas)):
        tank+=gas[i]-cost[i]

        if tank<0:
            start=i+1
            tank=0

    return start

gas=[1,2,4,5,9]
cost=[3,4,1,10,1]

gas_station(gas,cost)
```

## Complexity Analysis

- Time : $O(n)$
- Space : $O(1)$
