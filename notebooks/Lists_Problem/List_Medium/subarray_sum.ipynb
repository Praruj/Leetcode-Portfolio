# 560. Subarray Sum Equals K

Given an array of integers nums and an integer k, return the total number of subarrays whose sum equals to k.

A subarray is a contiguous non-empty sequence of elements within an array.


```python
def subarray(nums,k):
    count=0
    n=len(nums)
    for i in range(n):
        current_sum=0
        for j in range(i,n):
            current_sum+=nums[j]
            if current_sum==k:
                count+=1
    return count

nums=[9,4,20,3,10,5]
k=33
subarray(nums,k)
```

# Complexity analysis:
- Time complexity : O(n^2) because of nested loop iterating over the array to generate all possible subarrays.
- Space Compleity : O(1) because of extra data structure is used. 

```python
def subarray(nums,k):
    count=0
    current_sum=0
    store={0:1}

    for num in nums:
        current_sum+=num

        if (current_sum-k) in store:
            count+=store[current_sum-k]
        
        store[current_sum]=store.get(current_sum,0)+1
    
    return count

nums=[9,4,20,3,10,5]
k=33
subarray(nums,k)
```
