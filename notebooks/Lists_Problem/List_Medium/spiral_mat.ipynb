# 54. Spiral Matrix
Given an m x n matrix, return all elements of the matrix in spiral order.

## Solution
- First we found logic boundaries.
- While condition
- Corner case

```python
def spiralorder(matrix):
    if not matrix or not matrix[0]:
        return []

    m, n = len(matrix), len(matrix[0])
    srow, scol = 0, 0
    erow, ecol = m - 1, n - 1
    result = []

    while srow <= erow and scol <= ecol:
        # Top
        for j in range(scol, ecol + 1):
            result.append(matrix[srow][j])
        srow += 1

        # Right
        for i in range(srow, erow + 1):
            result.append(matrix[i][ecol])
        ecol -= 1

        # Bottom
        if srow <= erow:
            for j in range(ecol, scol - 1, -1):
                result.append(matrix[erow][j])
            erow -= 1

        # Left
        if scol <= ecol:
            for i in range(erow, srow - 1, -1):
                result.append(matrix[i][scol])
            scol += 1

    return result
matrix =[[1, 2, 3],[4, 5, 6],[7, 8, 9]]

spiralorder(matrix)
```
