# 74. Search a 2D Matrix

You are given an m x n integer matrix matrix with the following two properties:

Each row is sorted in non-decreasing order.
The first integer of each row is greater than the last integer of the previous row.
Given an integer target, return true if target is in matrix or false otherwise.

You must write a solution in `O(log(m * n))` time complexity.

```python
class Solution:
    def searchInRow(self, mat, target, row):
        cols = len(mat[0])
        start = 0
        end = cols - 1

        while start <= end:
            mid = start + (end - start) // 2
            if target == mat[row][mid]:
                return True
            elif target > mat[row][mid]:
                start = mid + 1
            else:
                end = mid - 1

        return False

    def searchMatrix(self, mat, target):
        rows, cols = len(mat), len(mat[0])
        start, end = 0, rows - 1

        while start <= end:
            mid = start + (end - start) // 2

            # check if target could be in this row
            if mat[mid][0] <= target <= mat[mid][cols - 1]:
                return self.searchInRow(mat, target, mid)

            # target is larger than the last element of this row → go down
            elif target > mat[mid][cols - 1]:
                start = mid + 1

            # target is smaller than the first element of this row → go up
            else:
                end = mid - 1

        return False
mat = [
    [1, 3, 5, 7],
    [10, 11, 16, 20],
    [23, 30, 34, 60]
]
target = 16

s = Solution()
print(s.searchMatrix(mat, target))   
```

# Time complexity analysis
The time complexity for both part of the code is O(logn) so total is also same.
