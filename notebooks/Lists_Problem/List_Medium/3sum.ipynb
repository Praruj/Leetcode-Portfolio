# 3Sum
Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

## Brute force solution

```python
def threesum(nums):
    n=len(nums)
    st=set()
    for i in range(0,n):
        for j in range(i,n):
            for k in range(j+1,n):
                if nums[i]+nums[k]+nums[j]==0:
                    triplet = tuple(sorted([nums[i], nums[j], nums[k]])) #converts to tuple (immutable, hashable for sets)
                    st.add(triplet)
    return list(st)
nums = [-1, 0, 1, 2, -1, -4]
threesum(nums)

# tuple() = "freeze list so set can use it.
```

## Approach with Hashing

```python
def threesum(nums):
    n=len(nums)
    result=[]
    for i in range(n-2):  # Don't need last 2 indices
        if i > 0 and nums[i] == nums[i-1]:  # Skip duplicate i's
            continue
            
        tar = -nums[i]  # nums[i] + nums[j] + nums[k] = 0
        s = set()
        

        for j in range(i+1,n):
            tofind=tar-nums[j]

            if tofind in s:
                triplet=sorted([nums[i],nums[j],tofind])
                result.append(triplet)

                while j+1<n and nums[j+1]==nums[j]:
                    j+=1
            s.add(nums[j])

    return result


nums = [-1, 0, 1, 2, -1, -4]
threesum(nums)
```

# Two Pointer Approach

```python
def threesum(nums):
    nums.sort()
    n=len(nums)
    ans=[]

     
    for i in range(n):
        if i>0 and nums[i]==nums[i-1]:
            continue    
        j=i+1
        k=n-1

        while j<k:
            sum=nums[i]+nums[j]+nums[k]
            if sum<0:
                j+=1
            elif sum>0:
                k-=1
            else:
                triplet=[nums[i],nums[j],nums[k]]
                ans.append(triplet)
                j += 1
                k -= 1

                while j<k and nums[j]==nums[j-1]:
                    j+=1
                while j < k and nums[k] == nums[k + 1]:
                    k -= 1
    return ans
                 
nums = [-1, 0, 1, 2, -1, -4]
threesum(nums)
```

## complexity analysis:

- Time Complexity = O(nlogn +n^2)
- Space complexity = O(triplet)
