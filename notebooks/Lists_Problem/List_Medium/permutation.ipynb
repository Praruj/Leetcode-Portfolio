# 46. Permutations
Given an array nums of distinct integers, return all the possible permutations. You can return the answer in any order

----
# Solution:
A permutation is every possible ordering of the elements.

Use Backtacking
- Build permutations one element at a time
- Try all valid choices
- Undo choices when needed and explore other options

```python
def ispermute(nums):
    result=[]

    def backtrack(path):
        #base case
        if len(path)==len(nums):
            result.append(path[:])
            return
        
        for num in nums:
            if num in path:
                continue

            path.append(num)
            backtrack(path)
            path.pop()
    backtrack([])
    return result
nums = [1, 2, 3]
ispermute(nums)
```

## Complexity Analysis

### 1. Time Complexity
Total work = (number of permutations) × (work per permutation)

- Number of permutations = `n!`
- Work per permutation = `O(n)`

So:
- `n! × O(n)`
- **Time Complexity = `O(n × n!)`**

---

### 2. Space Complexity
- Recursion stack space = `O(n)`
