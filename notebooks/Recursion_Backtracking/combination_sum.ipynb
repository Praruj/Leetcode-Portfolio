# Combination Sum
- array=[2,3,5] and target=8

```python
def comsum(nums,target):
    result=[]
    def backtrack(start,current,total):
        if total==target:
            result.append(list(current))
            return
        if total>target:
            return
        
        for i in range(start,len(nums)):
            current.append(nums[i])
            backtrack(i,current,total+nums[i])
            current.pop()
    backtrack(0,[],0)
    return result

nums=[2,3,5]
target=8
comsum(nums,target)
```

# Complexity Analysis
- Time complexity : O(N^(T\M+1))
- Space complexit: O(T/M)
