# Palindrome Partitioning


```python
def is_palindrome(s):
    return s==s[::-1]

def partition_helper(s,start,current_partition,result):
    if start==len(s):
        result.append(current_partition[:])
        return

    for end in range(start+1,len(s)+1):
        substring=s[start:end]


        if is_palindrome(substring):
            current_partition.append(substring)
            partition_helper(s,end,current_partition,result)
            current_partition.pop()

def palindrome_partitioning(s):
    result=[]
    partition_helper(s,0,[],result)
    return result

s = "aab"
print(palindrome_partitioning(s))
```

## Complexity analysis:

- Time complexity : O(2^n*n) .
- Space Complexity : O(2^n*n) .  

Space needed for recursion stack and storing results.
