# Rat in Maze

You are given a maze in the form of a matrix of size n * m. Each cell is either clear or blocked denoted by 1 and 0 respectively.
A rat sits at the top-left cell and there exists a block of cheese at the bottom-right cell. Both these cells are guaranteed to be clear.
You need to find if the rat can get the cheese if it can move only in one of the two directions - down and right. It can’t move to blocked cells.

```python
def helper(nums,r,c,path,ans,n):
        # Boundary + invalid checks
    if r < 0 or c < 0 or r >= n or c >= n or nums[r][c] == 0:
        return
        
        
    # Destination reached
    if r == n - 1 and c == n - 1:
        ans.append(path)
        return

    # Mark visited
    nums[r][c] = 0


    helper(nums,r+1,c,path+"D",ans,n) #Down
    helper(nums,r-1,c,path+"U",ans,n) #Up
    helper(nums,r,c-1,path+"L",ans,n) # left 
    helper(nums,r,c+1,path+"R",ans,n) # Right

def findpath(nums):
    n=len(nums)
    ans=[]

    if nums[0][0] == 0:
        return ans

    helper(nums, 0, 0, "", ans, n)
    return ans

maze = [
    [1, 0, 0, 0],
    [1, 1, 0, 1],
    [0, 1, 0, 0],
    [1, 1, 1, 1]
]

print(findpath(maze))
```

# Complexity Analysis:

- Time Complexity : O(4^(n^2))
    1. Each cell may lead to 4 recursive calls.
    2. Maximum recursion depth can be up to n^2. 
    so total recursive calls in worst case : 4^(n^2)

- Space Complexity: O(n^2)
    1. Recursion stack
    2. In worst case, number of paths can also be exponential.
