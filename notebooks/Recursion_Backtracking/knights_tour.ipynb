**Complexity Analysis**

Recursive function explores all possible paths in worst case.
- Each cell → 8 knight moves → each tries 8 moves → depth n²
- Total calls ≈ 8^(n²) in worst case
- Time Complexity : O(8^(n^2))

Space Complexity : O(n^2)
- Recursion Depth is up to n^2.
