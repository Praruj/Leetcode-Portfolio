# **The celebrity Problem**

Given a square matrix mat[][] of size n x n, where` mat[i][j] == 1 ` means person i knows person j, and` mat[i][j] == 0` means person i does not know person j, find the celebrity person where,

A celebrity is defined as someone who:

- Is known by everyone else does not know anyone (except themselves)
- Return the index of the celebrity if one exists, otherwise return -1.

```python
def celeb(nums):
    stack=[]
    n=len(nums)
    def knows(a, b):
        return nums[a][b] == 1
    stack=[]  
    for i in range(n):
        stack.append(i)
    
    while len(stack)>1:
        a=stack.pop()
        b=stack.pop()

        if knows(a,b):
            stack.append(b)
        else:
            stack.append(a)
    
    candidate=stack.pop()

    for i in range(n):
        if i!=candidate:
            if knows(candidate,i) or not knows(i,candidate):
                return -1
    return candidate  


nums = [
 [0, 1, 1],
 [0, 0, 1],
 [0, 0, 0]
]

celeb(nums)
```

## Complexity Analysis

- Time complexity : $O(n)$
- Space complexty: $O(n)$

**Two pointer approach**

```python
def celebrity(nums):
    n=len(nums)

    i=0
    j=n-1
    while i<j:
        if nums[i][j]==1: 
            i+=1
        else:
            j-=1
    c=i

    for i in range(n):
        if i==c:
            continue

        if nums[c][i] or not nums[i][c]:
            return -1
        
    return c


nums = [
 [0, 1, 1],
 [0, 0, 1],
 [0, 0, 0]
]

celebrity(nums)
```
