# valid parenthesis

- *Conditions*
1. Opening and closing brackets should be same.
2. open brackets must be closed in correct order.
3. Every close bracket has a corresponding open bracket of same type. 

```python
def isvalid(str):
    st=[]
    pairs={
        ')': '(',
        ']': '[',
        '}': '{'
    }

    for ch in str:
        if ch in pairs.values():
            st.append(ch)

        elif ch in pairs:
            if not st or st[-1]!=pairs[ch]:
                return False
            st.pop()
        else:
            return False

    return len(st)==0
(isvalid("()[]{}"))  
```

**Complexity Analysis**

- Time complexity : O(N)
    1. We traverse string once.
    2. Each character is processed in constant time.

- Space Complexity: O(N)
    1. In worst case, we have to store all the characters.

*Summary*:

Each element is pushed and popped at most once.

*without using dictionary*

```python
def isvalid(s):
    st = []
    
    for ch in s:
        if ch == '(' or ch == '[' or ch == '{':
            st.append(ch)
        elif ch == ')' or ch == ']' or ch == '}':
            if not st:
                return False
            top = st.pop()
            if ch == ')' and top != '(':
                return False
            elif ch == ']' and top != '[':
                return False
            elif ch == '}' and top != '{':
                return False
        else:
            return False 
    
    return len(st) == 0

print(isvalid("()[]{}"))  
```

```python
def isValid(s):  
    stack = []
    mapping = {')': '(', '}': '{', ']': '['}

    for char in s:
        if char in mapping:
            top = stack.pop() if stack else '#'
            if mapping[char] != top:
                return False
        else:
            stack.append(char)

    return not stack
```
