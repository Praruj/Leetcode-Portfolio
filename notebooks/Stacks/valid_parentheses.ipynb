# valid parenthesis

- *Conditions*
1. Opening and closing brackets should be same.
2. open brackets must be closed in correct order.
3. Every close bracket has a corresponding open bracket of same type. 

```python
def isvalid(str):
    st=[]
    pairs={
        ')': '(',
        ']': '[',
        '}': '{'
    }

    for ch in str:
        if ch in pairs.values():
            st.append(ch)

        elif ch in pairs:
            if not st or st[-1]!=pairs[ch]:
                return False
            st.pop()
        else:
            return False

    return len(st)==0
(isvalid("()[]{}"))  
```

**Complexity Analysis**

- Time complexity : O(N)
    1. We traverse string once.
    2. Each character ios processed in constant time.

- Space Complexity: O(N)
    1. In worst case, we have to store all the characters.

*Summary*:

Each elemnt is pushed and popped at most once.
