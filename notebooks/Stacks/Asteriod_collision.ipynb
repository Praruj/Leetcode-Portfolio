# Asteriod Collision

Collsion happens when a positive number comes before a negative number.
- [5, -3] → collision
- [-3, 5] → no collision


We use stack because we process asteriod one by one and need to compare the current asteriod with previous unresolved one 

```python
def asteriod_collision(asteriods):
    stack=[]

    for curr in asteriods:
        while stack and curr<0 and stack[-1]>0:
            if stack[-1]<-curr:
                stack.pop()
                continue

            elif stack[-1]==-curr:
                stack.pop()
            break
        else:
            stack.append(curr)

    return stack

asteriods=[3, 5, -2, -5, 8, -8]
asteriod_collision(asteriods)
```

## Complexity Analysis

- Time complexity : $O(n)$
- Space Complexity : $O(n)$
