# Stacks

A stack is bucket like Data structure, depends on  LIFO (last in first out).

*A collection of items with LIFO order*
- The last item we push (add) is the first one we pop (remove).

**How stacks are implemened**
- Arrays : Easy fix size.
- Linked Lists : Flexible size

Regardless of implementation, behaviour is always LIFO. 


**Common stacks patterns**

1. Reversing things
2. Undo Mechanisms
3. Balanced Symbols
4. DFS


**Stack Complexity**
- Push : O(1) - Just puts on top.
- Pop : O(1)  - Remove from top
- Search (non-top item): O(n) - In worst case, we have to scan everything.
- Peek : O(1) - inspect top



**Core operations**

1. push(x) : `stack.append(x)` -> Add item to top. 
2. pop() : `stack.pop()`   -> Removes top item.
3. peek()/top() : `stack[-1]` -> view top without removing
4. isEmpty() : `len(stack)==0` -> check if stack is empty.




---

```python
class Stack:
    def __init__(self):
        self.stack=[]

    def push(self,item):
        self.stack.append(item)
        print(f"pushed : {item}")

    def pop(self):
        if not self.is_empty():
            popped_item=self.stack.pop()
            print(f"popped: {popped_item}")
            return popped_item
        
        else:
            print("Empty Stack")
            return None

    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        else:
            print("Empty Stack")
            return None
        
    def is_empty(self):
        return len(self.stack)==0

    def size(self):
        return len(self.stack)

s=Stack()
s.push(20)
s.peek()
```

**Remove adjaced Elements**

```python
def remove_adjacent(s):
    stack=[]

    for char in s:
        if stack and stack[-1]==char:
            stack.pop()
        else:
            stack.append(char)
    return "".join(stack)

remove_adjacent("abbca")
```

**remove all adjacent duplicates that appear more than K times**

```python
def remove_duplicates(s, k):
    stack = []

    for char in s:
        if stack and stack[-1][0] == char:
            stack[-1] = (char, stack[-1][1] + 1)
            if stack[-1][1] > k:
                stack.pop()  # Remove the character if its count exceeds k
        else:
            stack.append((char, 1))  

    
    result = "".join(char * count for char, count in stack)
    return result


print(remove_duplicates("aabbccdd", 2))  
```
