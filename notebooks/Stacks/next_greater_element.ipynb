# Next Greater Element

```python
def greater(nums):
    stack=[]
    ans=[]

    for i in range(len(nums)-1,-1,-1):
        while len(stack)>0 and stack[-1]<=nums[i]:
            stack.pop()
        
        if len(stack)==0:
            ans.append(-1)
        
        else:
            ans.append(stack[-1])
        
        stack.append(nums[i])

    
    return ans[::-1]

greater([4, 5, 2, 10])
```

**Complexity Analysis**

- Time complexity : O(n)
- Space complexity : O(n)

*Another Variation of this question*

```python
def greater(nums1,nums2):
    map={} # pairs of nums2[i] and next greater
    stack=[]

    for i in range(len(nums2)-1,-1,-1):
        while len(stack)>0 and stack[-1]<=nums2[i]:
            stack.pop()

        if len(stack)==0:
            map[nums2[i]]=-1
        
        else:
            map[nums2[i]]=stack[-1]
        
        stack.append(nums2[i])
        
    ans=[]
    for i in range(len(nums1)):
        ans.append(map[nums1[i]])

    return ans
```
