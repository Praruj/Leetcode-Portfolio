# **Next Greater Element II**

`nums = [1,2,3,4,3]`

**Optimal approach**

```python
def next_greater(nums):
    n=len(nums)
    ans = [-1] * n     # Initialize with -1 as a default for elements with no greater neighbour 
    # If we never find a greater neighbor, it stays -1.

    stack=[]

    for i in range(2*n-1,-1,-1):
        while stack and stack[-1] <= nums[i % n]:
            stack.pop()

        if stack:
            ans[i%n]=stack[-1]
        stack.append(nums[i%n])

    return ans

n=[1,2,3,4,3]
next_greater(n)
```

*Logic used*:

1. By iterating from `2n-1` down to 0, we simulate the circular nature of array wihout doubling the memory. The modulo operator `%n` keeps our index within bounds.
2. We pop elements that are smaller than or equal to current number because they cant be "next greater" for anything to left of our current position.


**Complexity Analysis**

- Time complexity : $O(n)$ - Even though there is nested while loop each element is pushed and popped from stack exactly once.
- Space Complexity : $O(n)$ - To store the stack and result array.

## Brute force : 

With time complexity:  $O(n^2)$

```python
def brute_force(nums):
    n=len(nums)
    ans=[-1]*n

    for i in range(n):
        for j in range(i+1,i+1):
            if nums[j%n]>nums[i]:
                ans[i]=nums[j%n]
                break
    return ans
```
