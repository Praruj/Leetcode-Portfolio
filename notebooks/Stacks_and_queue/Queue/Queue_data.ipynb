# Queue

Queue (FIFO - First In First Out)

- Insertion: Enqueue (at rear)
- Deletion: Dequeue (from front)

Operations:
1. Enqueue(x) → Add element at rear → O(1)
2. Dequeue() → Remove element from front → O(1)
3. Front() → Get front element → O(1)
4. Rear() → Get last element → O(1)

Example:

Queue: [10, 20, 30]

Enqueue(40) → [10, 20, 30, 40]

Dequeue() → [20, 30, 40]

Types:
- Simple Queue
- Circular Queue
- Deque
- Priority Queue

----

**Implementation of queue**

1. Simple Code

```python
class Queue:
    def __init__(self):
        self.values=[]
    
    def enqueue(self,x):
        self.values.append(x)

    def dequeue(self):
        front=self.values[0]
        self.values=self.values[1:]
        return front
    
q1=Queue()
q1.enqueue(10)
q1.enqueue(20)
q1.enqueue(30)

q1.dequeue()
print(q1.values)
```

 **More advanced Code**

```python
from collections import deque

class Queue:
    def __init__(self):
        self.values=deque()

    def enqueue(self,x):
        self.values.append(x)
    
    def dequeue(self):
        if not self.is_empty():
            return self.values.popleft()
        return "Queue is empty"
    
    def front(self):
        if not self.is_empty():
            return self.values[0]
        return "Queue is empty"
    
    def is_empty(self):
        return len(self.values)==0

q=Queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

print("Front:", q.front())     
print("Dequeue:", q.dequeue()) 
print("Front:", q.front())     
```
