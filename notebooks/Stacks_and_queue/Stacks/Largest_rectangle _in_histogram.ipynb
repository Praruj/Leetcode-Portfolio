# **Largest Rectangle in Histogram**

**Brute force approach**

```python
def largest_rec(nums):
    n=len(nums)
    max_area=0


    for i in range(n):
        min_height=nums[i]
        for j in range(i,n):
            min_height=min(min_height,nums[j])
            width=j-i+1
            area=min_height*width
            max_area=max(max_area,area)
    return max_area
nums=[2,1,5,6,2,3]
largest_rec(nums)
```

**Complexity Analysis**

- Time complexity : $O(n^2)$

        Since, outer loop runs n times and inner loop also runs to n times. 

- Space complexity : $O(1)$

        We are using few variables without extra arrays, recursion and nothing growing with input size.

**Optimized approach**

On finding left smaller and right smaller value. 

```python
def largest_rec(nums):
    n = len(nums)
    stack = []

    left = [0] * n   # nearest smaller to left
    right = [0] * n  # nearest smaller to right

    # 1️⃣ Find nearest smaller to left
    for i in range(n):
        while stack and nums[stack[-1]] >= nums[i]:
            stack.pop()
        if stack:
            left[i] = stack[-1]
        else:
            left[i] = -1
        stack.append(i)

    # 2️⃣ Find nearest smaller to right
    stack = []
    for i in range(n-1, -1, -1):
        while stack and nums[stack[-1]] >= nums[i]:
            stack.pop()
        if stack:
            right[i] = stack[-1]
        else:
            right[i] = n
        stack.append(i)

    # 3️⃣ Calculate max area
    max_area = 0
    for i in range(n):
        width = right[i] - left[i] - 1
        area = nums[i] * width
        max_area = max(max_area, area)

    return max_area

# Test
nums = [2,1,5,6,2,3]
print(largest_rec(nums)) 
```

**Complexity Analysis**

- Time complexity: $O(n)$

        Each element is pushed/popped at most once.


- Space Complexity:  $O(n)$

        for left, right and stack.
