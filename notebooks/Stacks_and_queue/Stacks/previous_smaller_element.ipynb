# Previous smaller Elements

```python
def small_elment(nums):
    stack=[]
    ans=[]
    for i in range(len(nums)):
        while stack and stack[-1]>=nums[i]:
            stack.pop()

        if len(stack)==0:
            ans.append(-1)
        else:
            ans.append(stack[-1])

        stack.append(nums[i])
    return ans
nums = [9, 1, 8, 2, 7, 3, 6, 4, 5]
small_elment(nums)
```

## Complexity Analysis

- Time complexity : O(n)
    1. Each element is pushed onto the stack exactly once.
    2. Each element is poped from stack at most once

- Space Complexity : O(n)
    1. In worst case, all elements can stay in stack so size can go up to n.
    2. Ans stores one result per element.
