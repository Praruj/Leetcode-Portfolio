# Stock span

span= Maximum number of consecutives days from today -> `price<= todays price`

we are going to store index in stacks.

- prevhigh=span=i-prevhigh

```python
def span(price):
    stack=[]
    ans=[]

    for i in range(len(price)):
        while len(stack)>0 and price[stack[-1]]<=price[i]:  # when stack isnot empty and price at top of stack <= current price. 
            stack.pop()
        
        if len(stack)==0:
            ans.append(i+1)
        
        else:
            ans.append(i-stack[-1])
        stack.append(i)
    
    return ans
```

**Complexity Analysis**
- Time complexity : O(n)  Even though the while loop runs multiple times, it doesn’t restart from scratch each time.
- Space Complexity : O(n)   Both stack and ans can hold up to n elements.
