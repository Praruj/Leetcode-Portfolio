Given an array `arr[]` of distinct elements, which was initially sorted in ascending order but then rotated at some unknown pivot, the task is to find the index of a target key.  If the key is not present in the array, return -1.

- Input: arr[] = [5, 6, 7, 8, 9, 10, 1, 2, 3], key = 3
- Output: 8
- Explanation: 3 is found at index 8.

Constraints:
1. 1 ≤ arr.size() ≤ 106
2. 0 ≤ arr[i] ≤ 106
3. 0 ≤ key ≤ 106

## Approach:

We can solve this problem using linear search with a time complexity of `O(n)`. However, since the array is sorted, we should consider binary search, which is faster with a time complexity of `O(log n)`.

```python
def search(nums, k):
    low = 0
    high = len(nums) - 1

    while high >= low:
        mid = (high + low) // 2
        
        # Check if the middle element is the target
        if nums[mid] == k:
            return mid

        # To check if the left half is sorted
        if nums[low] <= nums[mid]:
            # If the key is within the left sorted portion
            if nums[low] <= k < nums[mid]:
                high = mid - 1
            else:
                low = mid + 1
        else:
            # If the key is within the right sorted portion
            if nums[mid] < k <= nums[high]:
                low = mid + 1
            else:
                high = mid - 1

    return -1  # If the key is not found

# Test the function
nums = [5, 6, 7, 8, 9, 10, 1, 2, 3]
k = 3
print(search(nums, k))  
```
