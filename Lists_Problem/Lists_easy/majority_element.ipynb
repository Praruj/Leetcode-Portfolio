# 169. Majority Element

Given an array `nums` of size n, return the majority element.
The majority element is the element that appears more than `⌊n / 2⌋` times. You may assume that the majority element always exists in the array.  

Example 1:
- `Input`: nums = [3,2,3]
- `Output`: 3

## Brute force solutions:

In brute force solution:
 - Outer Loop picks an element at position i. 
 - Scan the whole array (all j (inner loop)) and count how many are equal to that element.
 - if it count is more than half the length, return that suddenly. 

```python

def majority_elements(nums):
    n=len(nums)
    for i in range(n):
        count=0
        for j in range(n):
            if nums[j]==nums[i]:
                count+=1
        
        if count>n//2:
            return nums[i]
        
majority_elements([3,2,3])
```

### Time complexity of ths brute force solution:
1. Outer loop: i runs n operation.
2. Inner Loop: J runs 0,n-1 for each operation of i. so it runs n operations.
So the total operation is O(n^2) which will exceed the time limit of leetcode.

#### Space complexity:
Space wise, the algorithm only uses a few integer variables `(n,i,j,count)` so, O(1) (constant extra space.)

# Optimized solution

1. Sort array.
2.  Loop through sorted array and count consecutive elements; reset count when value changes.


i=1: nums[1]=2, current=2 → same → count=2
     Check 2 > 2? No

i=2: nums[2]=3, current=2 → different → reset: current=3, count=1  
     Check 1 > 2? No

i=3: nums[3]=3, current=3 → same → count=2
     Check 2 > 2? No

i=4: nums[4]=3, current=3 → same → count=3
     Check 3 > 2? YES → return 3
     

```python
def majority_elements(nums):
    n=len(nums)
    nums.sort()
    current=nums[0]
    count=1

    for i in range(1,n):
        if nums[i]==current:
            count+=1
        else:
            current=nums[i]
            count=1

        if count>n//2:
            return current
    return current
majority_elements([3,2,3])
```

## Time Complexity:
Sorting takes O(Nlogn) and loops takes O(n) operatons so combining these we get O(Nlogn) as time complexity. and Space complexity is same as brute force solution which is O(1) for extra variables.

## The Boyer-moore Majority Algorithm

The Boyer Moore majority vote algorithm is an algorithm for finding the majority of a sequence of elements using linear time and a Constant number of words of memory. 
If the majority exists, it will emerge as winner. Two variables candidate and count to keep track of the potential majority elements. 

```python
def majority_element(nums):
    freq=0
    ans=0
    
    for i in range(len(nums)):
        if freq==0:
            ans=nums[i]
        if ans==nums[i]:
            freq+=1
        else:
            freq-=1
    return ans
majority_element([1, 2, 1, 2, 1, 2, 1])
```

The algorithm is deleting pairs of different elements until only one survives.
