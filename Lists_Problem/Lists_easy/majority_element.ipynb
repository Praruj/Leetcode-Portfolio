# 169. Majority Element

Given an array `nums` of size n, return the majority element.
The majority element is the element that appears more than `⌊n / 2⌋` times. You may assume that the majority element always exists in the array.  

Example 1:
- `Input`: nums = [3,2,3]
- `Output`: 3

## Brute force solutions:

In brute force solution:
 - Outer Loop picks an element at position i. 
 - Scan the whole array (all j (inner loop)) and count how many are equal to that element.
 - if it count is more than half the length, return that suddenly. 

```python

def majority_elements(nums):
    n=len(nums)
    for i in range(n):
        count=0
        for j in range(n):
            if nums[j]==nums[i]:
                count+=1
        
        if count>n//2:
            return nums[i]
        
majority_elements([3,2,3])
```

### Time complexity of ths brute force solution:
1. Outer loop: i runs n operation.
2. Inner Loop: J runs 0,n-1 for each operation of i. so it runs n operations.
So the total operation is O(n^2) which will exceed the time limit of leetcode.

#### Space complexity:
Space wise, the algorithm only uses a few integer variables `(n,i,j,count)` so, O(1) (constant extra space.)
